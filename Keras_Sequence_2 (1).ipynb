```python
import keras
#making neural network

model = Sequential()
#input_layer
model.add(Dense(128, activation='relu',input_shape=(784, )))
#hidden_layer
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
#output layer

model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
#optimizer define
#loss function
adam=keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=12,batch_size=32,validation_data=(x_test,y_test))
# every epoc e kyta kore sample dibo = batch size
```

Why Add This Dimension?
Consistency: Keras models are designed to work with batches of data. Even if you're only using one image or data point, the model expects a batch. Adding an extra dimension keeps the input shape consistent.

Compatibility: During training or testing, you might use different numbers of samples. Adding the batch dimension ensures that your input format is always compatible with what the model expects, whether you're using one or multiple samples.

Avoid Errors: Without the correct input shape (including the batch dimension), Keras will raise an error because it won't be able to match the input data to the model's expected input shape.

So, in summary, you add the extra dimension to make sure your input data matches what the Keras model expects, ensuring smooth processing without errors.
